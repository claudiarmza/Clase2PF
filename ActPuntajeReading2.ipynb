```python
import pandas as pd
import numpy as np

data_students=pd.read_csv("C:\\Users\\CRAMIREZA\\Downloads\\clean_students_complete.csv")
data_students
```

```python
#1. Crea / Presenta la forma que creas más conveniente para describir qué grado por escuela tuvo el mejor puntaje en "reading", 
#y con ello saber que genero es el que predomina en  este ﬁltro.
```

```python
data_students.set_index("Student ID")
data_students=data_students.drop(["Unnamed: 0"],axis=1)
data_students.head(3)
```

```python
#Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el peor puntaje
#  en "reading", y con ello saber que genero es el que predomina en este ﬁltro.

max_reading=data_students[data_students["reading_score"]==data_students["reading_score"].max()]
max_reading_corto=max_reading[["gender","school_name","grade"]]
dummiesgen1=pd.get_dummies(max_reading_corto[["gender"]])
max_reading_base=pd.concat([max_reading_corto,dummiesgen1],axis=1)
max_reading_base=max_reading_base.groupby(["grade","school_name"]).sum()
max_reading_base
```

```python
#Como son 15 escuelas se va a colocar [0:15] dependiendo del grado que se quiera ver y se hará una gráfica para cada grado.
```
