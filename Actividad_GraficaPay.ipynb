```python
import pandas as pd
import numpy as np

datos=pd.read_csv("C:\\Users\\CRAMIREZA\\Downloads\\athlete_events.csv")
datos
```

```python
datosMedal=datos[datos["Medal"].notna()]
datosMedal
```

```python
actividad=datos[["Year","Season","Medal","NOC"]]
```

```python
recent_year=datos[datos["Year"]==2016]
```

```python
recent_year
```

```python
dummies=pd.get_dummies(recent_year[["Medal"]])
dummies.head()
```

```python
concatenado=pd.concat([recent_year,dummies],axis=1)
```

```python
agrupado=concatenado.groupby(["Year","Medal"]).sum()
agrupado.head(10)
```

```python
agrupado["Total_Medals"]=agrupado["Medal_Gold"]+agrupado["Medal_Silver"]+agrupado["Medal_Bronze"]
agrupado.head(6)
```

```python
nombres=["Bronce","Oro","Plata"]
pay=agrupado["Total_Medals"][:3].plot.pie(figsize=(10,10), labels=nombres,autopct="%0.1f %%")
```

```python
fig=pay.get_figure()
```

```python
fig.savefig("GráficaPay")
```
